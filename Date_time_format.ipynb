```python
import pandas as pd

# Sample data similar to your input
data = {
    'Opened': [
        '6/13/2024 11:42:31 AM', '6/13/2024 11:41:46 AM', '6/13/2024 9:30:00 AM',
        '6/13/2024 9:30:00 AM', '6/13/2024 9:20:42 AM', '06-12-2024 16:51',
        '06-12-2024 16:50', '06-12-2024 16:48', '06-12-2024 16:48',
        '06-12-2024 16:48', '06-12-2024 14:11', '06-12-2024 12:23'
    ]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Convert the 'Opened' column to datetime
df['Opened'] = pd.to_datetime(df['Opened'], errors='mixed')

# Format the datetime to 'YYYY-MM-DD HH:MM'
df['Opened'] = df['Opened'].dt.strftime('%Y-%m-%d %H:%M')

# Display the result
print(df)
```

```python
import pandas as pd
from datetime import datetime, timedelta

# Sample data
df = pd.read_excel('PT data.xlsx')
# df['Opened'] = df['Opened'].astype(str)
def custom_date_format(date_str):
    if isinstance(date_str, datetime):
        return date_str.strftime("%Y-%m-%d %H:%M")

    try:
        return datetime.strptime(date_str, "%m/%d/%Y %I:%M:%S %p").strftime("%Y-%m-%d %H:%M")
    except ValueError:
        pass

    try:
        return datetime.strptime(date_str, "%Y-%d-%m %H:%M:%S").strftime("%Y-%m-%d %H:%M")
    except ValueError:
        pass

    return date_str

# Apply the custom date format function
df['Opened'] = df['Opened'].apply(custom_date_format)
df['Opened'] = pd.to_datetime(df['Opened'])
# df['Opened'] = pd.to_datetime(df['Opened'], format='mixed')

# Display the result
df[(df['Opened'].dt.month >= 6) & (df['Opened'].dt.year == 2024)]
```

```python
# prompt: read PT data.xlsx file here in opened there are different types of date groupby the df with respect to type in 'Opened', dont convert to pd.to_datetime

import pandas as pd

# Read the PT data.xlsx file
df = pd.read_excel('PT data.xlsx')

# Create a new column to store the type of each value in 'Opened'
df['Opened_type'] = df['Opened'].apply(lambda x: type(x).__name__)

# Group by the type of values in the 'Opened' column
grouped_df = df.groupby('Opened_type')

grouped_dfs = {name: group.drop(columns=['Opened_type']) for name, group in grouped_df}

# Print the groups
for name, group in grouped_df:
    print(f"Group: {name}")
    # print(group)
```

```python
string_df = grouped_dfs.get('str')
string_df['Opened'] = pd.to_datetime(string_df['Opened'], format='%m/%d/%Y %I:%M:%S %p')
string_df['Opened'] = string_df['Opened'].dt.strftime('%Y-%m-%d %H:%M:%S')
string_df['Opened'] = pd.to_datetime(string_df['Opened'])
string_df
```

```python
datetime_df = grouped_dfs.get('datetime')
datetime_df['Opened'] = pd.to_datetime(datetime_df['Opened'])
datetime_df['Opened'] = datetime_df['Opened'].dt.strftime('%Y-%d-%m %H:%M:%S')
datetime_df
```

```python
merged_df = pd.concat([string_df, datetime_df], ignore_index=True)
merged_df['Opened'] =  pd.to_datetime(merged_df['Opened'])
merged_df
```

```python
# prompt: from merged_df give alerts from june month only

june_alerts = merged_df[merged_df['Opened'].dt.month == 6]
june_alerts
```
